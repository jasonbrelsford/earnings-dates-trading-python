```python
!pip install yfinance pandas numpy pandas_market_calendars matplotlib requests

import warnings
warnings.filterwarnings("ignore")

from datetime import datetime, timedelta
import pandas as pd
import numpy as np
import yfinance as yf
import pandas_market_calendars as mcal
import matplotlib.pyplot as plt
import requests

# ---- Configuration ----
TICKERS = [
    "AAPL","MSFT","GOOGL","AMZN",
    "META","AVGO","AMD","CRM","NFLX",
    "ORCL","INTC","TSLA","ADBE","QCOM"
]
YEARS = 5
END_DATE = datetime.utcnow().date()
START_DATE = END_DATE - timedelta(days=int(YEARS*365.25))
# Replace with your actual API key
# API_KEY = "sCPefVMepA6AGADELHhAhd60lzjmmf" # FMP API key is not needed if using yfinance
# ------------------------

def nyse_trading_days(start, end):
    nyse = mcal.get_calendar("XNYS")
    sched = nyse.schedule(start_date=start, end_date=end)
    return pd.DatetimeIndex(mcal.date_range(sched, frequency="1D").tz_convert(None))

def prev_day(trading_days, dt):
    ts = pd.Timestamp(dt).normalize() + pd.Timedelta(hours=16)
    d = trading_days[trading_days < ts]
    return None if d.empty else d[-1]

def next_day(trading_days, dt):
    ts = pd.Timestamp(dt).normalize()
    d = trading_days[trading_days >= ts]
    return None if d.empty else d[0]

def get_earnings_dates(ticker, start, end):
    # Using a placeholder for earnings dates as FMP earnings data is not readily available in the free tier
    print(f"[WARN] {ticker}: using approx quarterly schedule for earnings dates")
    return pd.date_range(start=start, end=end, freq="QS")

# Function to get historical prices using yfinance
def get_historical_prices_yfinance(ticker, start, end):
    try:
        # Fetch data using yfinance. Download method returns a DataFrame
        data = yf.download(ticker, start=start, end=end)
        if not data.empty:
            # Print the columns of the fetched DataFrame
            print(f"Columns fetched for {ticker}: {data.columns.tolist()}")
            return data
    except Exception as e:
        print(f"[ERROR] Could not fetch data for {ticker} from yfinance: {e}")
        return pd.DataFrame()
    return pd.DataFrame()


def run_strategy(ticker, trading_days, start, end): # Removed api_key parameter
    print(f"Running strategy for {ticker}")
    # Use yfinance to get historical prices
    prices = get_historical_prices_yfinance(ticker, start - pd.Timedelta(days=7), end + pd.Timedelta(days=7))
    if prices.empty:
        print(f"No price data for {ticker}")
        return []

    print(f"Fetched {len(prices)} price points for {ticker}")
    earnings = get_earnings_dates(ticker, start, end)
    earnings = earnings[(earnings >= prices.index.min()) & (earnings <= prices.index.max())]
    print(f"Found {len(earnings)} earnings dates for {ticker}")

    trades=[]
    for i in range(len(earnings)-1):
        b = next_day(trading_days, earnings[i]+pd.Timedelta(hours=16))
        s = prev_day(trading_days, earnings[i+1])
        print(f"Earnings date {earnings[i].date()}: Potential buy date {b}, Potential sell date {s}")
        if not b or not s or s<=b:
            print(f"Skipping trade for earnings date {earnings[i].date()}: Invalid buy/sell dates or sell date before buy date")
            continue

        # Ensure the dates are timezone-naive and remove the time component
        b = b.tz_localize(None).normalize() if b is not None and b.tzinfo else b.normalize() if b is not None else None
        s = s.tz_localize(None).normalize() if s is not None and s.tzinfo else s.normalize() if s is not None else None

        print(f"Prices index type: {type(prices.index)}, Buy date type: {type(b)}, Sell date type: {type(s)}")
        print(f"Buy date in prices index: {b in prices.index}, Sell date in prices index: {s in prices.index}")


        if b not in prices.index or s not in prices.index:
            print(f"Skipping trade for earnings date {earnings[i].date()}: Buy date {b.date()} or sell date {s.date()} not in price index")
            continue
        # Extract the scalar value from the pandas Series
        buy=prices.loc[b,"Open"].iloc[0]
        sell=prices.loc[s,"Close"].iloc[0]
        # Include ticker in the trades list
        trades.append((ticker, b.date(),buy,s.date(),sell,(sell/buy)-1))
        print(f"Trade generated for earnings date {earnings[i].date()}: Buy {b.date()} at {buy}, Sell {s.date()} at {sell}, Return {(sell/buy)-1}")
    return trades

trading_days = nyse_trading_days(START_DATE, END_DATE)
portfolio=[]
for tk in TICKERS:
    print("Running",tk)
    # Call run_strategy without the api_key
    t=run_strategy(tk,trading_days,START_DATE,END_DATE)
    if t: portfolio.extend(t)

# Update DataFrame columns to include 'Ticker'
df=pd.DataFrame(portfolio,columns=["Ticker","BuyDate","BuyPrice","SellDate","SellPrice","Return"])

if not df.empty:
  # Calculate cumulative returns and portfolio value
  df["Cumulative_Return"] = (1+df["Return"]).cumprod()
  # Assuming an initial investment of 1000 for the cumulative return calculation
  INITIAL_INVESTMENT = 1000
  df["Total_Portfolio_Value"] = df["Cumulative_Return"] * INITIAL_INVESTMENT

  # Calculate cumulative winning trades
  df["Winning_Trades"] = (df["Return"] > 0).cumsum()

  df.to_csv("trades.csv",index=False)

  # Plot total portfolio equity against trade number (default index)
  plt.plot(df.index, df["Total_Portfolio_Value"])
  plt.title("Total Portfolio Value Over Trades")
  plt.xlabel("Trade Number")
  plt.ylabel("Total Portfolio Value")
  plt.show()

  print("Total return:",df["Total_Portfolio_Value"].iloc[-1] - INITIAL_INVESTMENT)
else:
  print("No trades were generated.")
```

## Modify the `run strategy` function

### Subtask:
Update the `run_strategy` function to accept `sell_days_before` and `buy_days_after` parameters to allow for variations in trading strategy around earnings dates.


**Reasoning**:
The subtask requires modifying the `run_strategy` function to accept parameters for adjusting buy and sell dates relative to earnings dates. This involves changing the function signature and updating the logic for calculating `b` and `s` within the function, using the provided `prev_day` and `next_day` helper functions to ensure trading days are used.



```python
def run_strategy(ticker, trading_days, start, end, sell_days_before=0, buy_days_after=0):
    print(f"Running strategy for {ticker} with sell_days_before={sell_days_before}, buy_days_after={buy_days_after}")
    # Use yfinance to get historical prices
    # Fetch data for a wider range to ensure we have data for buy/sell dates around earnings
    prices = get_historical_prices_yfinance(ticker, start - pd.Timedelta(days=30), end + pd.Timedelta(days=30))
    if prices.empty:
        print(f"No price data for {ticker}")
        return []

    print(f"Fetched {len(prices)} price points for {ticker}")
    earnings = get_earnings_dates(ticker, start, end)
    # Filter earnings dates to be within the fetched price data range
    earnings = earnings[(earnings >= prices.index.min()) & (earnings <= prices.index.max())]
    print(f"Found {len(earnings)} earnings dates for {ticker}")

    trades=[]
    for i in range(len(earnings)-1):
        # Calculate buy date: buy_days_after trading days after current earnings
        b_candidate = earnings[i] + pd.Timedelta(hours=16) # Start checking after the earnings announcement
        b = b_candidate
        for _ in range(buy_days_after):
            b = next_day(trading_days, b)
            if b is None:
                break # Stop if no next trading day is found

        # Calculate sell date: sell_days_before trading days before next earnings
        s_candidate = earnings[i+1] - pd.Timedelta(hours=16) # Start checking before the earnings announcement
        s = s_candidate
        for _ in range(sell_days_before):
            s = prev_day(trading_days, s)
            if s is None:
                break # Stop if no previous trading day is found

        print(f"Earnings date {earnings[i].date()}: Potential buy date {b}, Potential sell date {s}")

        if not b or not s or s<=b:
            print(f"Skipping trade for earnings date {earnings[i].date()}: Invalid buy/sell dates or sell date before buy date")
            continue

        # Ensure the dates are timezone-naive and remove the time component
        b = b.tz_localize(None).normalize() if b is not None and b.tzinfo else b.normalize() if b is not None else None
        s = s.tz_localize(None).normalize() if s is not None and s.tzinfo else s.normalize() if s is not None else None

        print(f"Prices index type: {type(prices.index)}, Buy date type: {type(b)}, Sell date type: {type(s)}")
        # Check if calculated buy and sell dates are in the price index
        if b not in prices.index or s not in prices.index:
            print(f"Skipping trade for earnings date {earnings[i].date()}: Buy date {b.date() if b else 'N/A'} or sell date {s.date() if s else 'N/A'} not in price index")
            continue

        # Extract the scalar value from the pandas Series
        buy=prices.loc[b,"Open"].iloc[0]
        sell=prices.loc[s,"Close"].iloc[0]

        # Include ticker in the trades list
        trades.append((ticker, b.date(),buy,s.date(),sell,(sell/buy)-1))
        print(f"Trade generated for earnings date {earnings[i].date()}: Buy {b.date()} at {buy}, Sell {s.date()} at {sell}, Return {(sell/buy)-1}")
    return trades
```

## Create a list of strategy variations

### Subtask:
Define a list of tuples, where each tuple represents a strategy variation with specific sell-before-earnings and buy-after-earnings day offsets. Include the original strategy, selling two days before and buying one day after, and selling the day of and buying the day after.


**Reasoning**:
Define the list of strategy variations as tuples of (sell_days_before, buy_days_after).



```python
strategy_variations = [
    (0, 0),  # Original strategy: sell on earnings day, buy on earnings day (or next trading day after earnings if earnings is after market close)
    (2, 1),  # Sell 2 days before earnings, buy 1 day after earnings
    (0, 1),  # Sell on earnings day, buy 1 day after earnings
    (1, 1), # Sell 1 day before earnings, buy 1 day after earnings
    (1, 0) # Sell 1 day before earnings, buy on earnings day (or next trading day after earnings)
]

print("Defined strategy variations:")
for sell_days, buy_days in strategy_variations:
    print(f"Sell {sell_days} days before earnings, Buy {buy_days} days after earnings")
```

## Run backtests for each variation

### Subtask:
Iterate through the list of strategy variations, call the modified `run_strategy` function with the appropriate parameters for each variation, and store the results (the DataFrame of trades) for each variation.


**Reasoning**:
Iterate through the defined strategy variations, run the trading strategy for each ticker and variation, and store the resulting trades in a dictionary.



```python
all_trades = {}

for sell_days, buy_days in strategy_variations:
    strategy_name = f"Sell {sell_days} Buy {buy_days}"
    print(f"\n--- Running backtest for strategy: {strategy_name} ---")
    portfolio_variation = []
    for tk in TICKERS:
        t = run_strategy(tk, trading_days, START_DATE, END_DATE, sell_days_before=sell_days, buy_days_after=buy_days)
        if t:
            portfolio_variation.extend(t)

    df_variation = pd.DataFrame(portfolio_variation, columns=["Ticker", "BuyDate", "BuyPrice", "SellDate", "SellPrice", "Return"])
    all_trades[strategy_name] = df_variation

print("\nBacktesting for all strategy variations complete.")
```

## Calculate cumulative portfolio value for each variation

### Subtask:
For each strategy variation's trade DataFrame stored in the `all_trades` dictionary, calculate the cumulative portfolio value over time.


**Reasoning**:
Iterate through the all_trades dictionary and calculate the cumulative portfolio value for each strategy's DataFrame.



```python
INITIAL_INVESTMENT = 1000

for strategy_name, df_variation in all_trades.items():
    if not df_variation.empty:
        df_variation["Cumulative_Portfolio_Value"] = (1 + df_variation["Return"]).cumprod() * INITIAL_INVESTMENT
        print(f"Calculated cumulative portfolio value for strategy: {strategy_name}")
    else:
        print(f"No trades found for strategy: {strategy_name}. Skipping cumulative value calculation.")

print("\nCumulative portfolio value calculation for all strategies complete.")
```

## Plot cumulative portfolio values

### Subtask:
Plot the cumulative portfolio value over time for all strategy variations on a single graph to visually compare their performance.


**Reasoning**:
Plot the cumulative portfolio value for each strategy variation on a single graph to compare their performance visually, according to the instructions.



```python
plt.figure(figsize=(14, 7))

for strategy_name, df_variation in all_trades.items():
    if not df_variation.empty:
        plt.plot(df_variation.index, df_variation["Cumulative_Portfolio_Value"], label=strategy_name)

plt.title("Cumulative Portfolio Value for Different Trading Strategies")
plt.xlabel("Trade Number")
plt.ylabel("Portfolio Value")
plt.legend()
plt.grid(True)
plt.show()
```
